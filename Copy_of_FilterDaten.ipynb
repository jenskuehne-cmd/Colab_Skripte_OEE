```python
# @title
import matplotlib.pyplot as plt
from collections import Counter

# Mindestanzahl für Rollen als Variable definieren
min_count = 20

# Rollen aus der Spalte 'S4H_Roles' extrahieren und zählen
all_roles = []
for roles in combined_df['S4H_Roles']:
    if isinstance(roles, str):
        all_roles.extend([role.strip() for role in roles.split(',') if role.strip()])

role_counts = Counter(all_roles)

# Rollen filtern, die mindestens 'min_count'-mal vorkommen
filtered_roles = {role: count for role, count in role_counts.items() if count >= min_count}

# Daten für das Diagramm vorbereiten
sorted_roles = sorted(filtered_roles.items(), key=lambda x: x[1], reverse=True)
roles, counts = zip(*sorted_roles)

# Balkendiagramm erstellen
plt.figure(figsize=(12, 8))
bars = plt.barh(roles, counts, color='skyblue')
plt.xlabel('Anzahl', fontsize=14)
plt.ylabel('S4H Roles', fontsize=10)
plt.title(f'Häufigkeit der S4H Rollen (mindestens {min_count} Vorkommen)', fontsize=16)
plt.gca().invert_yaxis()  # Reihenfolge umkehren, damit die häufigste Rolle oben ist
plt.yticks(fontsize=8)  # Schriftgröße der Rollennamen anpassen

# Balken beschriften
for bar, count in zip(bars, counts):
    plt.text(count + 1, bar.get_y() + bar.get_height() / 2, str(count), va='center', fontsize=10)

# Diagramm anzeigen
plt.tight_layout()
plt.show()
```

## Zusammenfassung des Notebooks

Dieses Notebook dient dazu, Rollen- und Organisationsdaten aus verschiedenen Quellen zu konsolidieren, zu analysieren und zu visualisieren. Es führt die folgenden Hauptschritte aus:

1.  **Datenimport und Zusammenführung:**
    *   Es liest Benutzer- und Rollendaten aus drei Google Sheets (KAU, PTD, PS) ein.
    *   Die S4H-Rollen werden bereinigt und pro Benutzer in einer Spalte zusammengeführt.
    *   Alle Daten werden in einem einzigen DataFrame (`combined_df`) zusammengeführt.

2.  **Datenexport:**
    *   Die kombinierten Daten werden als Excel- und CSV-Dateien exportiert.

3.  **Rollen- und Organisationsanalyse:**

```python
# @title
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display

# --- Cluster-Definition ---
clusters = {
    "Technical Unit": [
        'Technical Unit', 'Value Stream Vials Speed Technical Unit FG',
        'Technik', 'Technical Unit E2E', 'Solids Launch & Supply Technical Unit DP',
        'Solids Launch & Supply Technical Unit FG'
    ],
    "QA": [
        'QA Solids', 'QA for QC', 'QA Vials Speed', 'QA Vials Volume', 'QA Rocephin', 'QA for Functions',
        'QA Syringes & Devices', 'QA Logistics',
        'Quality Systems', 'Quality Kaiseraugst', 'Quality Compliance', 'Internal E2E Quality Operations',
        'QA for Functions B', 'IMP Quality Operations Basel&Kau', 'QS QV'
    ],
    "QC": [
        'QC Release & ID K', 'QC Release & ID St', 'BBK QC', 'Sample Mgmt BS', 'QC Support BS & KAU',
        'ASAT', 'QC Sample Mgmt KAU', 'QC mEMT Basel', 'QC mEMT Kaiseraugst', 'Bioassay',
        'Microbiology', 'Non-sterile Materials & Lab-Sterilisation Processes', 'SM HPLC, KF, Disso'
    ],
    "Produktion": [
        'Rocephin DP Fill', 'Process Unit FG', 'Process Unit Visual Inspection',
        'Kaiseraugst Drug Product Manufacturing', 'Solids Launch & Supply Blister 2 FG',
        'Process Unit DP', 'Solids Launch & Supply Production DP', 'Vials Speed',
        'SLS Blister 1 FG', 'Blister', 'Solids Launch & Supply Blister 1 FG',
        'Rocephin Manufacturing KAU', 'Vials Volume', 'Process Unit Fertigspritzen',
        'Fertigspritzen & Assemblierung', 'Solids Launch & Supply',
        'Solids Launch & Supply Print Service & Support'
    ],
    "ES&T": [
        'Utilities', 'Maintenance Operations', 'Reliability',
        'Unit Lead Asset Process Validation 1', 'Unit Lead Asset Process Validation 2',
        'Enabling Science & Operational Support', 'Engineering Science and Technology',
        'CoE Reinigungsvalidierung', 'Monitoring'
    ]
}

def generate_heatmap(selected_cluster_name):
    # --- Daten vorbereiten ---
    df = combined_df.copy()

    if 'User - Organization' not in df.columns or 'S4H_Roles' not in df.columns:
        print("Error: combined_df must contain 'User - Organization' and 'S4H_Roles' columns.")
        return

    if selected_cluster_name not in clusters:
        print(f"Error: Unknown cluster '{selected_cluster_name}'. Available: {list(clusters.keys())}")
        return

    orgs = clusters[selected_cluster_name]

    filtered = df[df['User - Organization'].isin(orgs)].copy()
    if filtered.empty:
        print(f"No data found for cluster '{selected_cluster_name}'.")
        return

    # Rollen explodieren
    filtered = filtered.assign(Role=filtered['S4H_Roles'].str.split(',')).explode('Role')
    filtered['Role'] = filtered['Role'].astype(str).str.strip()

    # Falls leere Rollen existieren
    filtered = filtered[filtered['Role'].ne('') & filtered['Role'].ne('nan')]

    # --- Matrix: Y=Role, X=Abteilung (User - Organization), Wert=Anzahl ---
    value_col = 'UserID' if 'UserID' in filtered.columns else 'Role'  # fallback, falls UserID fehlt
    role_dept_matrix = pd.pivot_table(
        filtered,
        index='Role',
        columns='User - Organization',
        values=value_col,
        aggfunc='count',
        fill_value=0
    )

    # Sortieren: Spalten nach Summe, Rollen nach Summe
    role_dept_matrix = role_dept_matrix.loc[
        role_dept_matrix.sum(axis=1).sort_values(ascending=False).index,
        role_dept_matrix.sum(axis=0).sort_values(ascending=False).index
    ]

    # --- Plot: Heatmap mit Zahlen und kleinerer Schrift ---
    plt.figure(figsize=(16, 10)) # Anpassen der Figure-Größe für bessere Lesbarkeit
    sns.heatmap(
        role_dept_matrix.T, # Transponieren, damit Abteilungen auf der Y-Achse sind
        annot=True, # Zahlen in den Zellen anzeigen
        fmt='d', # Format für die Zahlen (Ganze Zahlen)
        cmap='viridis', # Farbschema anpassen (optional)
        linewidths=.5, # Linien zwischen den Zellen
        linecolor='black', # Farbe der Linien
        annot_kws={"fontsize": 6}, # Schriftgröße der Annotationen
        cbar_kws={'label': 'Anzahl Rollen'}
    )

    plt.xticks(rotation=90, fontsize=6) # 90° gedreht für Rollen auf X-Achse, kleinere Schrift
    plt.yticks(rotation=0, fontsize=8) # 0° gedreht für Abteilungen auf Y-Achse, kleinere Schrift

    plt.title(f'Rollen pro Abteilung im Cluster: {selected_cluster_name}', fontsize=14)
    plt.xlabel('Rollen', fontsize=10)
    plt.ylabel('Abteilungen (User - Organization)', fontsize=10)
    plt.tight_layout()
    plt.show()

# --- Dropdown-Widget erstellen ---
cluster_selector = widgets.Dropdown(
    options=list(clusters.keys()),
    value=list(clusters.keys())[0], # Set default value to the first cluster
    description='Select Cluster:',
    disabled=False,
    style={'description_width': 'initial'}
)

# --- Interaktive Ausgabe erstellen und anzeigen ---
interactive_output = widgets.interactive_output(generate_heatmap, {'selected_cluster_name': cluster_selector})

display(cluster_selector, interactive_output)
```
